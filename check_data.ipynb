```python
import glob
```

```python
MIRROR = "/global/cfs/cdirs/m3522/cmip6/CMIP6"
VARIANT = "r1i1p1f1"

checks = [
    ("MPI-ESM1-2-HR", "MPI-M",     "CMIP",        "historical"),
    ("MPI-ESM1-2-HR", "MPI-M",     "ScenarioMIP", "ssp370"),
    ("GFDL-ESM4",     "NOAA-GFDL", "CMIP",        "historical"),
    ("GFDL-ESM4",     "NOAA-GFDL", "ScenarioMIP", "ssp370"),
]

for model, inst, activity, exp in checks:
    for var in ["tasmax", "tasmin", "hurs"]:
        patt = f"{MIRROR}/{activity}/{inst}/{model}/{exp}/{VARIANT}/day/{var}/*/*/*.nc"
        n = len(glob.glob(patt))
        print(f"[{'OK ' if n else '-- '}] {model:14s} {exp:11s} {var:6s}: {n:3d} files")
```
